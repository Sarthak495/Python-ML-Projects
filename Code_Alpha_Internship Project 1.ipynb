```python
# Visulization
```

```python
plt.hist(data,bins=20)
plt.title("Histogram",size=15)
plt.xlabel("Region")
plt.ylabel("Area")
plt.show()
```

```python
sns.barplot(x=data["Area"],y=data["Estimated Labour Participation Rate (%)"],ci=None,estimator=np.max,hue=data["Area"])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=data["Area"],y=data["Estimated Labour Participation Rate (%)"],hue=data["Area"])
plt.show()
```

```python
sns.distplot(data["Estimated Unemployment Rate (%)"])
```

```python
sns.stripplot(x=data["Area"],y=data["Estimated Employed"],hue=data["Area"])
```

# Sesonal trends in data

```python
sns.boxplot(data["Estimated Unemployment Rate (%)"])
plt.title('Monthly Seasonal Unemployment Rate Trends by Region')
plt.xlabel('Month')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.tight_layout()
plt.show()
```
